-  ### Subtask 2.2: 

#### Here we are dropping the columns which we do not need further in our analysis. ( 1 mark )

```python
movies = movies.drop(['color', 
                      'director_facebook_likes', 
                      'actor_3_facebook_likes', 
                      'actor_1_facebook_likes', 
                      'cast_total_facebook_likes', 
                      'actor_2_facebook_likes', 
                      'duration', 
                      'facenumber_in_poster', 
                      'content_rating', 
                      'country', 
                      'movie_imdb_link', 
                      'aspect_ratio',
                      'plot_keywords',
                      'actor_2_name',
                      'actor_3_name'], 
                       axis = 1)
```

#### Here we are removing the duplicate rows if any.

```python
movies.drop_duplicates(subset = None, keep = 'first', inplace = True)
```

-  ### Subtask 2.3:

#### In the below cell we are removing entire rows having null values in 'gross' and 'budget' columns. ( 1 mark )

```python
movies = movies[~np.isnan(movies['gross'])]
movies = movies[~np.isnan(movies['budget'])]
movies
```

-  ### Subtask 2.4: 

#### In the below cell we are considering the rows having null values less than 5. ( 1 mark )

```python
movies = movies[movies.isnull().sum(axis=1) <= 5]
movies
```

-  ### Subtask 2.5: 

#### In the below cell we are checking for null values in language column, and if found update it with 'English'. ( 1 mark )

```python
movies['language'].describe()
movies.loc[pd.isnull(movies['language']), ['language']] = 'English'
movies
```

## Task 3: Data Analysis

-  ### Subtask 3.1: Convert units of gross and budget to millions for all movies in the dataset. (2 marks)

```python
# Write your code here
movies['gross'] = movies['gross']/1000000
movies['budget'] = movies['budget']/1000000
movies.head()
```

-  ### Subtask 3.2:  Display the 10 movies with the highest profit. (Blockbuster Movies playlist) (2 marks)

```python
# Write your code here
movies1 = movies.copy()
movies1['movie_title'] = movies1['movie_title'].apply(lambda x:x.strip())
movies1['Profit'] = movies1['gross'] - movies1['budget']
movies1 = movies1.sort_values(by=['Profit'], ascending = False)
blockbuster_movie_list = movies1['movie_title'].head(10).to_list()
print(blockbuster_movie_list)
```

-  ### Subtask 3.3: Display the top 10 movies (based on IMDb ratings) with a minimum of 25,000 ratings (Most Loved movies playlist) (4 marks)

```python
# Write your code here
movies1 = movies.copy()
movies1['movie_title'] = movies1['movie_title'].apply(lambda x:x.strip())
movies1 = movies1[movies1['num_voted_users'] >= 25000]
movies1 = movies1.sort_values(by=['imdb_score'], ascending = False) 
loved_movie_list = movies1['movie_title'].head(10).to_list()
print(loved_movie_list)
```

- ### Subtask 3.4: Display the names of  the top 5 directors based on IMDb ratings with a minimum of 5 movies (Top Directors Playlist)  (4 marks)

```python
# Write your code here
movies1 = movies.copy()
director_df = movies1.groupby('director_name').agg({'movie_title':'count','imdb_score':'mean'})
top_director_playlist = director_df.loc[director_df['movie_title'] >= 5].sort_values(by=['imdb_score'], ascending=False)
top_director_playlist.head(5).index.to_list()
```

-  ### Subtask 3.5.1: Find the most popular genre of movies (by highest average rating) (4 marks)

```python
# Write your code here
movies1 = movies.copy()
movies1['genres_1'] = movies1['genres'].apply(lambda x:x.split('|')[0])
primary_genre_imdb = movies1.groupby('genres_1')['imdb_score'].agg('mean')
highest_imdb_genre = primary_genre_imdb.sort_values(ascending=False).idxmax()
highest_imdb_genre
```

-  ### Subtask 3.5.2: Find the most popular genre of movies (by total gross) (4 marks)

```python
# Write your code here
movies1 = movies.copy()
movies1['genres_1'] = movies1['genres'].apply(lambda x:x.split('|')[0])
primary_genre_gross = movies1.groupby('genres_1')['gross'].agg('sum')
highest_gross_genre = primary_genre_gross.sort_values(ascending=False).idxmax()
highest_gross_genre
```
